```python
# Compound Interest Program

# input values
P = float(input("Enter Principal: "))
R = float(input("Enter Rate: "))
n = float(input("Enter number of times interest compounded per year: "))
t = float(input("Enter Time in years: "))

# calculation
A = P * (1 + R/n) ** (n * t)

# output
print("Compound Amount =", A)
print("Data Type of Amount is", type(A))
```
