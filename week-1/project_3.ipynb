```python
# Annuity Plan Program

# input values
PMT = float(input("Enter payment each period: "))
R = float(input("Enter interest rate: "))
n = float(input("Enter number of payments per year: "))
t = float(input("Enter time in years: "))

# calculation
A = PMT * ((1 + R/n) ** (n*t) - 1) / (R/n)

# output
print("Future Value of Annuity =", A)
print("Data Type of Amount is", type(A))
```
