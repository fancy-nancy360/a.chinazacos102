```python
Start
Input years of experience.
Input age of staff.
If experience > 25 AND age ≥ 55
ATR = 5,600,000
Else if experience > 20 AND age ≥ 45
ATR = 4,480,000
Else if experience > 10 AND age ≥ 35
ATR = 1,500,000
Else
ATR = 550,000
Display ATR
Stop
```

```python
# Input data
experience = int(input("Enter years of experience: "))
age = int(input("Enter age: "))

# Determine ATR
if experience > 25 and age >= 55:
    atr = 5600000
elif experience > 20 and age >= 45:
    atr = 4480000
elif experience > 10 and age >= 35:
    atr = 1500000
else:
    atr = 550000

# Output
print("Annual Tax Revenue (ATR): ₦", atr)
```
